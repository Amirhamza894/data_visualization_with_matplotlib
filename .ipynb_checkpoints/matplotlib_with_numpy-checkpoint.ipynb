## Methods

First Method

```python
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

Second Method

```python
fig = plt.figure()
ax = fig.add_axes([1,2,3,4])
ax.plot(x,y)
plt.show()
```

Third Method

```python
fig, ax = plt.subplots()
ax.plot(x,y);
```

## Matplotlib in workflow

```python
import matplotlib.pyplot as plt

x =[2,3,4,5,6]
y = [11,22,33,44,55]

fig, ax = plt.subplots(figsize=(3,3))

ax.plot(x,y)

ax.set(title = 'Simple Plot',
      xlabel = 'X axis',
      ylabel = 'Y label')

plt.show()
```

## Making plots with Numpy arrays with the help of Matplotlib

Plots
 * line Plot
 * Scatter Plot
 * Bar Plot
 * Histogram
 * subplots

```python
import numpy as np
from numpy.random import random
np.random.seed(0)
```

```python
x = np.linspace(0,10, 50)
x[:10]
```

```python
plt.plot(x);
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
plt.scatter(x, x**10);
```

```python
fig, ax = plt.subplots()
ax.scatter(x, x**10);
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
fruits_price = {'Banana':10,
               'Apple':15,
               'Pinapple':12}
fig, ax = plt.subplots()
ax.bar(fruits_price.keys(), fruits_price.values());
ax.set(title='Fruits Price',
      ylabel='Price ($)');
```

```python
fig, ax = plt.subplots()
ax.barh(list(fruits_price.keys()), list(fruits_price.values()));
```

```python
x = np.random.randn(1000)

fig, ax = plt.subplots()
ax.hist(x, 40);
```

### Two types of subplots

#### Option One

```python
fig, ((ax, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5));

ax.plot(x, x/2);
ax2.scatter(np.random.random(100), np.random.random(100));
ax3.bar(fruits_price.keys(), fruits_price.values());
ax4.hist(x, 30);
```

#### Option Two

```python
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5));
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(100), np.random.random(100));
ax[1, 0].bar(fruits_price.keys(), fruits_price.values());
ax[1,1].hist(x, 30);
```
